```python
def acc(database, negative = False):
    ones = 0
    for i in range(len(database)):
        pred = model(database[i][0].view(-1,1,779))
        pred = np.argmax(pred.detach().numpy())
        if pred == database[i][]:
            ones+=1
    acc = ones/len(database) 
    if negative:
        acc = 1-acc
    return acc
```

```python
#get true positive of model use noise-free data, acc = 0.9758861827827345
acc(npc)
```

```python
#get true positive of model use 'full of 0 data', acc = 0.18808777429467086
acc(noise_zero_npc)
```

```python
#true positive of model use fixed spec(from all zero), acc = 0.9746804919218712 
acc(output_npc)
```

```python
#true negative of model use fixed spec(from all zero), acc = 0.3664772727272727

acc(output_control,negative = True)
```

```python
#true negative of model noise free control data, acc = 0.373083475298126

acc(control,negative = True)
```
